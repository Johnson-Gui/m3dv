```python
import numpy as np
from matplotlib import pyplot as plt
import os
```

```python
tpath='./train_val'
file_num=0
if not os.path.isdir(tpath):
    print('false')
else:
    for item in os.listdir(tpath):
        if not file_num:
            tmp=np.load(os.path.join(tpath,item))
            voxel= tmp['voxel']/255
            mask=tmp['seg']
            voxel=np.multiply(voxel,mask)
            voxel=voxel.reshape(1,100,100,100,1)
            file_num+=1
        else:
            tmp=np.load(os.path.join(tpath,item))
            voxel1= tmp['voxel']/255
            mask=tmp['seg']
            voxel1=np.multiply(voxel1,mask)
            voxel1=voxel1.reshape(1,100,100,100,1)
            voxel = np.concatenate((voxel,voxel1), axis = 0)
            print(os.path.join(item))
            file_num+=1
print(file_num)
```
